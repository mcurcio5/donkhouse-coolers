```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N = 200
l = .03
X1 = 20

estimated_Ns = []

for i in range(100000):
    variables = np.sort(np.random.exponential(scale=1/l, size=N))[:X1]
    l2 = X1 / np.sum(variables)
    estimated_Ns.append(l2/l * X1)
```

```python
sns.set_style('white')
sns.set_context('notebook')
```

```python
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(12)
sns.histplot(data = estimated_Ns, bins = 100, ax = ax)
ax.set_title('Bootstrapped Estimation for true N=200 and X1=20', fontsize=18)
ax.set_xlabel('N', fontsize = 16)
ax.set_ylabel('Count', fontsize = 16)
plt.savefig('bootstrapped.png', format='png')
plt.show()
```
