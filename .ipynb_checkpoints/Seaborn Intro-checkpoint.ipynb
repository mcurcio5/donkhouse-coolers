```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('COVID-19 Activity.csv')
```

```python
df['REPORT_DATE'] = df['REPORT_DATE'].apply(lambda d: datetime.strptime(d, '%Y-%m-%d'))
```

```python
df.head()
```

```python
weekly_cases_by_country = df[['PEOPLE_POSITIVE_NEW_CASES_COUNT', 'REPORT_DATE', 'COUNTRY_SHORT_NAME', 'CONTINENT_NAME']].groupby([
    'CONTINENT_NAME', 'COUNTRY_SHORT_NAME', pd.Grouper(key='REPORT_DATE', freq = '1w')]).sum().reset_index()
```

```python
len(set(weekly_cases_by_country['COUNTRY_SHORT_NAME']))
```

```python
colors = plt.cm.get_cmap('Spectral')(np.linspace(0,1,5))
```

```python
# if log scale
weekly_cases_by_country['PEOPLE_POSITIVE_NEW_CASES_COUNT'] = weekly_cases_by_country['PEOPLE_POSITIVE_NEW_CASES_COUNT'].apply(lambda x: x+1)
```

```python
weekly_cases_by_country.groupby('CONTINENT_NAME').groups.keys()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style

# style.use('fivethirtyeight')
sns.set_theme()
sns.set_style('whitegrid')
sns.set_context('notebook')


groups = weekly_cases_by_country.groupby('CONTINENT_NAME')

continents = list(groups.groups.keys())
n_continents = len(continents)

fig, axs = plt.subplots(n_continents)

sns.despine(fig, left=True, bottom=True)
fig.set_figheight(8*n_continents)
fig.set_figwidth(8)

for i, ax in enumerate(axs.flat):
    ax.set_title('Weekly Cases for %s' % (continents[i]))
    ax.set_xlabel('Date', fontsize = 10)
    ax.set_ylabel('Number of Weekly Cases', fontsize = 10)
    
    continent_groups = groups.get_group(continents[i]).groupby('COUNTRY_SHORT_NAME')
    
    n_continent_groups = len(continent_groups.groups.keys())
    colors = plt.cm.get_cmap('tab20b')(np.linspace(0, 1, n_continent_groups))
    
    for j, (country, group) in enumerate(continent_groups):
        ax.plot(group.REPORT_DATE, group.PEOPLE_POSITIVE_NEW_CASES_COUNT, label = country, color = colors[j])
    
    #ax.legend(loc = 'best', ncol = 3, fontsize = 10)
    #ax.set_yscale('log')

plt.show()
```
