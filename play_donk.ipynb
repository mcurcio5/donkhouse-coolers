```python
def start_up_table(table_link, user):
    # window1 login
    room_id = int(table_link[-5:])
    driver = webdriver.Chrome('/Users/mcurcio/Downloads/chromedriver')
    driver.get('https://donkhouse.com/groups')
    # login
    try:
        e=WebDriverWait(driver, max_delay).until(EC.presence_of_element_located((By.NAME, 'name')))
        e=WebDriverWait(driver, max_delay).until(EC.presence_of_element_located((By.NAME, 'password')))
        e=WebDriverWait(driver, max_delay).until(EC.presence_of_element_located((By.XPATH, '//button[text()="Login"]')))
    except TimeoutException:
        raise Exception('logging in took too much time')
        
    sleep(.1)
    
    username = 'mcurcio' if user == 'mcurcio' else '-you'
    password = 'qyjtid-3vingi-coSpaj' if user == 'mcurcio' else '54321donk'
    
    driver.find_element_by_name('name').send_keys(username)
    sleep(.1)
    driver.find_element_by_name('password').send_keys(password)
    
    driver.find_element(By.XPATH, '//button[text()="Login"]').click()
    
    sleep(1)
    
    driver.get(table_link)
    
    seat = 6 if user == 'mcurcio' else 4
    sleep(1)
    declare_game_functions(driver, seat, room_id)
    sleep(1)
    sit_and_set_up(driver, seat, room_id)
    sleep(2)
```

```python
def stand(driver, seat, room_id):
    command ='socket.emit("action", {type: "stand", sizing: 0, refill: 0, hand_id: "mUmLm6yh6CoEKXr0", seat: ' + str(
        seat) + ', room_id: '  + str(room_id) + ', action_id: 114})'
    driver.execute_script(command)
```

```python
def add_chips(driver, seat, room_id, n_chips):
    command ='socket.emit("action", {type: "add", sizing: 0, refill: ' + str(
        n_chips * 100) + ', hand_id: "mUmLm6yh6CoEKXr0", seat: ' + str(
        seat) + ', room_id: '  + str(room_id) + ', action_id: 114})'
    driver.execute_script(command)
```

```python
def sit_in(driver, seat, room_id):
    command ='socket.emit("action", {type: "sit in", sizing: 0, refill: 0, hand_id: "mUmLm6yh6CoEKXr0", seat: ' + str(
        seat) + ', room_id: '  + str(room_id) + ', action_id: 114})'
    driver.execute_script(command)
```

```python
def declare_game_functions(driver, seat, room_id):
    driver.execute_script("""
        var button;
        var is_preflop;
        var my_seat = """ + str(seat) + """;
        var room_id = """ + str(room_id) + """;
        var current_hand_id;
        var current_action_id;
        var current_gamestate;
        var my_chip_stack;
        var current_seat_to_act;
        var cards_dealt;
        
        socket.on("game starts", (data) => {
            button = data.game.button;
            current_seat_to_act = button;
            current_hand_id = data.game.hand_id;
            is_preflop = true;
            cards_dealt = false;
        });

        socket.on("gamestate", (data) => {
            console.log(data.game.hand_id)
            current_hand_id = data.game.hand_id;
        });

        socket.on("resetting for next hand", (data) => {
            button = data.button;
            current_hand_id = data.hand_id;
            console.log(current_hand_id)
            is_preflop = true;
            cards_dealt = false;
        });

        socket.on("all holecards", () => {
            cards_dealt = true;
            setTimeout(function() {
                showHoleCards();
            }, 4000)
            if (current_seat_to_act == my_seat) {
                setTimeout(function() {
                    betFlop();
                }, 4000)
            }
        });

        socket.on("dealing to board", () => {
            is_preflop = false;
        });
        
        function sitIn() {
            socket.emit("action", {type: "sit in", sizing: 0, refill: 0, hand_id: current_hand_id, 
                seat: my_seat, room_id: room_id, action_id: current_action_id});
        }
        
        function showHoleCards() {
            socket.emit("action", {type: "show cards at end of hand", value: true, seat: my_seat,
                room_id: room_id, hand_id: current_hand_id});
        }
        
        function betFlop() {
            socket.emit("action", {type: "raise", sizing: 10000, refill: 0, 
                hand_id: current_hand_id, seat: my_seat, room_id: room_id, action_id: current_action_id});
        }

        function callFlop() {
            socket.emit("action", {type: "call", sizing: 9900, refill: 0, hand_id: current_hand_id, 
                seat: my_seat, room_id: room_id, action_id: current_action_id});
        }

        function Check() {
            socket.emit("action", {type: "check", sizing: 100, refill: 0, hand_id: current_hand_id, 
                seat: my_seat, room_id: room_id, action_id: current_action_id});
        }

        function addChips(n_chips) {
            socket.emit("action", {type: "add", sizing: 0, refill: n_chips, hand_id: current_hand_id, 
                seat: my_seat, room_id: room_id, action_id: current_action_id});
        }
        
        socket.on("requesting action", (data) => {
            current_action_id = data.action_id;
            current_seat_to_act = data.seat;
            
            if (current_seat_to_act != my_seat) {}
            else if (!is_preflop) {
                setTimeout(function() {
                    Check();
                }, 1500);
            }
            else if (button != my_seat) {
                setTimeout(function() {
                    callFlop();
                }, 1500);
            }
        });
        
        socket.on("distributing pot", (data) => {
            my_chip_stack = data.stack_consistency_check[""" + str(seat) + """];
            if (my_chip_stack == 0) {
                addChips(70000 - my_chip_stack);
                setTimeout(function() {
                    sitIn();
                }, 3000);
            }
            else if (my_chip_stack < 70000) {
                addChips(70000 - my_chip_stack);
            }
        });
        
        socket.on("player timed out", (data) => {
            console.log(data)
            if (data == my_seat) {
                setTimeout(function() {
                    sitIn();
                }, 2000);
            }
        });
        
        socket.on("game options push", (data) => {
            if (!data.current_player_is_sitting_in) {
                setTimeout(function() {
                    sitIn();
                }, 2000);
            }
        });
        
    """)
```

```python
def sit_and_set_up(driver, seat, room_id):
    sit(driver, seat, room_id)
    sleep(2)
    add_chips(driver, seat, room_id, 1000)
    sleep(2)
    sit_in(driver, seat, room_id)
    sleep(1)
```
